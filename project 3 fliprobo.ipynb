# Q3

```python
product_urls = []
start = 0
end = 3
for page in range(start,end):
    link = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in link[0:10]:
        product_urls.append(i.get_attribute("href"))
    nxt_button = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    nxt_button.click()
    time.sleep(2)                # scrpping all urls
```

```python
Brand = []
for link in product_urls[:30]:
    driver.get(link)
    
    try:
        brand = driver.find_element(By.XPATH, '/html/body/div[2]/div[2]/div[5]/div[4]/div[4]/div[44]/div/table/tbody/tr[1]/td[2]/span')
        Brand.append(brand.text)
    except NoSuchElementException:
        Brand.append('-')          # scrapping brand name from urls
```

```python
product_name = []
for link in product_urls[:30]:
    driver.get(link)
    
    try:
        name = driver.find_element(By.XPATH, '//span[@class="a-size-large product-title-word-break"]')
        product_name.append(name.text)
    except NoSuchElementException:
        product_name.append('-')                   # scrapping product name from url and handling exceptions with a dash
```

```python
Price =[]
for link in product_urls[:30]:
    driver.get(link)
    
    try:
        price = driver.find_element(By.XPATH, '//span[@class="a-price-whole"]')
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('-')                          #scrapping price name from url and handling exceptions with a dash
```

```python
Delivery =[]
for link in product_urls[:30]:
    driver.get(link)
    
    try:
        deli = driver.find_element(By.XPATH, '//span[@class="a-text-bold"]')
        Delivery.append(deli.text)
    except NoSuchElementException:
        Delivery.append('-')
```

```python
Availability =[]
for link in product_urls[:30]:
    driver.get(link)
    
    try:
        avail = driver.find_element(By.XPATH, '//span[@class="a-size-medium a-color-success"]')
        Availability.append(avail.text)
    except NoSuchElementException:
        Availability.append('-')
```

```python
Return =[]
for link in product_urls[:30]:
    driver.get(link)
    
    try:
        ret = driver.find_element(By.XPATH, '//a[@class="a-size-small a-link-normal a-text-normal"]')
        Return.append(ret.text)
    except NoSuchElementException:
        Return.append('-')
```

```python
guitar_df = pd.DataFrame({'Brand':Brand,'Name of Product':product_name,'Price':Price,'Delivery':Delivery,'Availability':Availability,'Return':Return,'Urls':product_urls})
guitar_df    #final scrapped data in a dataframe
```
